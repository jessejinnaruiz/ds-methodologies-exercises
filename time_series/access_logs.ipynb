### Concat columns 5-9 into a single column. This column represents the requests made

```python
log_data['Request']= log_data[5]+log_data[6].astype(str)+log_data[7].astype(str)+log_data[8]+log_data[9]
log_data.head()
```

```python
log_data = log_data.drop([1, 2, 4, 5, 6, 7, 8, 9], axis=1)
```

```python
log_data = log_data.rename(columns={0:'IP_address',3: 'date_time'})
```

```python
log_data.head()
```

```python
log_data['date_time'] = log_data['date_time'].str.replace('[','')
```

```python
log_data.head()
```

```python
log_data['date_time'] = log_data['date_time'].str.replace(':',' ',n=1)
```

```python
log_data.head()
```

```python
log_data['date_parsed'] = pd.to_datetime(log_data['date_time'], format = "%d/%b/%y %H%M%S", infer_datetime_format=True)
```
